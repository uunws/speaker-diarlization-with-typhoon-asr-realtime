```python
import yt_dlp
import os

def download_and_convert_to_wav(youtube_url, output_name="audio_test"):
    ydl_opts = {
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'wav',
            'preferredquality': '192',
        }],
        'outtmpl': 'temp_audio', # ชื่อไฟล์ชั่วคราว
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([youtube_url])

    final_filename = f"{output_name}.wav"
    os.system(f"ffmpeg -i temp_audio.wav -ar 16000 -ac 1 -y {final_filename}")

    # ลบไฟล์ชั่วคราว
    if os.path.exists("temp_audio.wav"):
        os.remove("temp_audio.wav")

    print(f"✅ แปลงไฟล์สำเร็จ! ไฟล์ของคุณคือ: {final_filename}")
    return final_filename

# --- ใช้งานจริง ---
url = "https://www.youtube.com/watch?v=Sjwkhso9h8U"
audio_file = download_and_convert_to_wav(url, "my_medical_test")
```

```python
# Load model directly
from transformers import AutoProcessor, AutoModelForSpeechSeq2Seq

processor = AutoProcessor.from_pretrained("typhoon-ai/typhoon-whisper-turbo")
model = AutoModelForSpeechSeq2Seq.from_pretrained("typhoon-ai/typhoon-whisper-turbo")
```

```python
import torch
from transformers import pipeline

dtype = torch.bfloat16 if torch.cuda.is_available() else torch.float32
device = "cuda:0" if torch.cuda.is_available() else "cpu"

pipe = pipeline(
    "automatic-speech-recognition",
    model=model,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    dtype=dtype,
    device=device,
)
```

```python
result = pipe(
    audio_file,
    chunk_length_s=30,
    stride_length_s=5,
    generate_kwargs={
        "language": "thai",
        "task": "transcribe"
    },
    return_timestamps=True
)

print(result["text"])
```

ใส่เกลือเพิ่มให้แบบว่าให้มันอร่อยถูกป่ะ เขาเลยมีงานวิจัยว่าจริงจริงแล้วอ่ะ น้ำส้มแก้ว้วนึงอะ น้ำตาลเท่าเท่ากับแบบน้ำอารมณ์หนึ่งกระป๋องเลย ก็คือประมาณแบบยี่สิบกรัม ซึ่งวันหนึ่งอะเราบอกว่าคุณสามารถกินน้ำตาลได้แค่ยี่สิบสี่กรัม per day ทั้งวันเลยนะ เพราะว่าคุณกินน้ำส้มแก้วนึงก็คือแบบว่าอันอื่นคุณไม่ต้องกินแล้วอย่างเงี้ย เออมันมันก็ไม่โอเคใช่ไหม ฉะนั้นก็เลยบอกว่าแทนที่เราจะกินแบบว่ากินน้ำส้มอย่างเงี้ย อาจจะกินเป็นส้มแทน เพะการกินส้มอ่ะคุณต้องเคี้ยวใช่ป่ะเคี้ยวแล้วก็กว่าจะกลืนกลืนเสร็จแล้วเนี่ยยังมีพวกไฟเบอร์อะไรต่าง ต่างอีกอะไรอย่างเงี้ยมันก็เลยเป็นการเหมือนกับว่าทำให้ได้สารอาหารครบกว่าแล้วก็ได้พลังงานที่น้อยกว่าอันนี้สำคัญมากแต่ว่าจริง จริงแล้วเนี่ยการที่เราจะเลือกการกินน่ะมันไม่ใช่แค่แบบว่าเฮ้ยจะเป็นจากแบบเขาเรียกอะไรเป็นจากน้ำส้มเป็นแค่ส้มเฉย เฉยอ่ะเราต้องมีการศึกษาเพิ่มเติมอีกนิดนึงว่าเฮ้ยอาหารอาหารหรือว่าผลไม้ประเภทไหนที่มีดัชดีน้ำตาลต่ำนิดนึงดัชดีน้ำตาลคืออะไรหมายถึงว่าพอเรากินเข้าไปแล้วเนี่ยร่างกายจะดูดซึมอ่ะแล้วกลายเป็นน้ำตาลในกระแสเลือดได้เร็วมากน้อยแค่ไหนเช่นถ้าสมมุติว่าจิมมี่มาถึงอ่ะจิมมี่กินอะไรกินกินสับปะรดพอจิมมี่กินสับปะรดแป๊บนึงอ่ะน้ำตามันขึ้นปึ๊ดเลยเร็วมากอย่างเงี้ยแทนกับการหมอฮิมี่บอกว่าขอกินฝรั่งแทนถ้าการกินฝรั่งอย่างเงี้ยก็ทำให้น้ำตาลอ่ะมันขึ้นช้ากว่าทั้ง ทั้งที่แคลอรี่อาจจะเท่ากันนะแต่น้ำตาลมันน้อยกว่าค่อนข้างเยอะอืออันนี้ก็เลยบอกว่าถ้าสมมุติว่าเราจะเลือกการกินน้ำผลไม้สมมุติต้องกินเป็นน้ำผลไม้จริง จริงอ่ะอันดับแรกคือต้องศึกษาก่อนว่าผลไม้แบบไหนควรเอามาทำเป็นน้ำผลไม้เช่นถ้าแบบว่าสับปะรดก็อาจจะเลี่ยงนิดนึงเลิกผลไม้ที่มีน้ำตาลต่ำใช่อาจจะเลิกเป็นแบบฝรั่งไหมเป็นแก้วมังกรไหมเป็นเบอร์รี่อะไรต่าง ต่างอะไรอย่างเงี้ยที่มีความเปรี้ยวนิดนึงอย่างเงี้ยคือโอเคอันนี้คืออันที่หนึ่งเนาะอันที่สองก็คือพยายามเลี่ยงอะไรพวกเขาเรียกว่านะเป็นน้ำผลไม้ที่สกัดกากอ่ะอืมเพราะว่าถ้าคุณเอากากออกคุณเหลือแต่น้ำอ่ะคุณก็คือเหลือแต่น้ำตาลกับน้ำตาลเลยใช่คือเหลือคือแบบว่าดื่มน้ำตาลเพียวเพียวฉะนั้นการที่เรากินน้ำผลไม้อ่ะจริง จริงเป็นการปั่นอาจจะโอเคกว่าอ่ะเนาะอันนี้คืออันที่สองอันที่สามที่สำคัญมากอ่ะคือคุณไม่ควรใ่อะไรปรุงแต่งเพิ่มแล้ว เช่นแบบว่าคนไทยชอบกินแบบใส่น้ำผึ้ง น้ำเชื่อมอะไรอย่างเงี้ย คิดว่ามันโอเค แต่จริง จริงแล้วอ่ะ คุณอาจจะไม่แบบว่าลืมนับไง ว่าจริง จริงแล้ววันหนึ่งอ่ะคุณกินน้ำตาลเยอะเกินไปหรือเปล่า ฉะนั้นถ้าเรากินประมาณเนี้ยก็คิดว่าน่าจะโอเค เมื่อกี้เราก็พูดถึงเรื่องการนับแคลไปแล้วเนาะ คิโตเจนิกไปแล้วเนี่ย แล้วก็การดื่ม เอ่อ จูส นะครับ น้ำผลไม้ เค้านี่อีกอันนึงก็คือ High Protein Diet หลายคนอาจจะคิดว่า เออ กินโปรตีนหนึ่งพันแคล กิน Fat หนึ่งพันแคล หรือการกินแป้งหนึ่งพันแคลอ่ะ มันเท่ากันหมด แล้วอันไหนดีที่สุด คือจริงจริงอ่ะ พอเรามาดูแล้วเนี่ย การกินโปรตีนหนึ่งพันแคลอ่ะ มันจะทำให้อิ่มนานมากกว่า เปรียบเทียบกับ Fat กับ Carbohydrate อือ เพราะอะไร เพราะการกินโปรตีน โปรตีนเนี่ยเป็นมอลิคุณใหญ่ แล้วใช้เวลาในการย่อยนาน นานกว่าจะย่อยในกระเพาะการจะได้ในลำไส้ แล้วมันจะเปลี่ยนตัวเองเนี่ยกลายเป็นกรดอามิโนแอซิด ลองคิดดูเรากินโปรตีนเข้าไปย่อยในอาหารของเราแล้วร่างกายเราจะดูดซึมโปรตีนตัวเนี้ยเปลี่ยนเป็นอามิโนแอซิดเข้าหลอดเลือด เข้ากระแสเลือด กระแสเลือดพอกระแสเลือดส่งไปในอัยวะต่าง ต่างใวร่างกายเราจะดูดซึมโปรตีนตัวนี้เปลี่ยนเป็นอามิโนซิสเข้าหลอดเลือด เข้ากระแสเลือด กระแสเลือด พอกระแสเลือดส่งไปในอัยวะต่าง ต่างในร่างกาย ดูดซึมเข้าไปอัยวะอื่น อือ มันต้องเปลี่ยนตัวเองจากกดอามิโนซิสมาเป็นโปรตีน โปรตีนอีกแล้วตอนขั้นตอนทุกขั้นตอนน่ะมันใช้พลังงานหมดเลย อือ หลายคนอ่ะก็เลยเอาว่าเออ การกินโปรตีนอ่ะเป็นวิธีหนึ่งที่มันเผาผ่านได้ค่อนข้างดี และเราเราทำให้เราอิ่มนานมากเดิมด้วย เพราะมลิกุลมันใหญ่นั่นเอง ก็แปลว่าสมมุติว่าเรากินในเทียบเท่ากับพลังงานเท่าเทียมกันของอาหารสามอย่างโปรตีนแฟตอย่างเงี้ยกับคราฟใช่ป่ะ ก็คือการกินโปรตีนคิดว่าน่าจะอ้วนน้อยสุด อ้วนน้อยสุดเพราะว่าแฟตกับคราฟเนี่ยกินปุ๊บย่อยปุ๊บเข้าเลย อ๋อใช้ได้เลยใช่ ใช้ได้เลยเปลี่ยนแปแปลได้เลย แต่ว่าคราวนี้เราต้องมาดูกันว่าเอ๊ะแล้วเราจะเลือกโปรตีนกินยังไง แพนเบสหรือ Animal base ก็ก็ขึ้นกับว่าคนไข้ของเราอยากกินแบบไหนอันไหนเหมาะกับตัวเขาแล้วก็ที่สำคัญเนี่ยถ้าใครอยากคิดมากินท้ายแพนเบสโปรตีนต้องระวังเรื่องของโรคไตให้มาปรึกษาแพทย์ก่อนละกันแล้วค่อยว่าเออควรกินโปรตีนเท่าไหร่และแบบไหนอืมเพราะว่าจริง จริงแล้วมันก็สำคัญนะเพราะว่าจริง จริงแล้วแบบว่าในยุคปัจจุบันอ่ะเรามีแบบว่าชอยที่เลือกกินเยอะมากใช่ป่ะไม่ว่าจะเป็นแพลนเบสเอ่ยหรือเป็นเวย์อะไรอย่างเงี้ยก็เลยอยากจะรู้ว่าจริง จริงแล้วสำหรับคนไข้คนนึงอ่ะเขาควรเลือกกินอะไรมากกว่าคือถ้าสำหรับลดน้ำหนักนะอืมคือจริง จริงลดน้ำหนักเนี่ยแพลนเบสจะดีกว่าอืมเพราะมันทำให้เรารู้สึกอิ่มมากกว่า คือถ้าสำหรับลดน้ำหนักนะ อือ คือจริงจริงลดน้ำหนักเนี่ย Plant based จะดีกว่า อือ เพราะมันทำให้เรารู้สึกอิ่มมากกว่า Animal based อิ่มมากกว่าอย่างเดียวหรืออิ่มนานด้วย อิ่มมากกว่าและอิ่มนานด้วย เมื่อเทียบกับ Normal weight protein หรือ Animal based protein อ่า แต่ว่าถ้าสมมุติว่าคนไข้คนนั้นเนี่ยมีโรคหรือมีอะไรอย่างเงี้ย ก็คือมันสามารถกิน Plant based ได้ทุกคนเลยป่ะ ก็คราวนี้เนี่ยอย่างหมอก็มีคนไข้ที่เป็นโรคไทรอยด์เนาะ ไทรอยด์ต่ำ ไทรอยด์ต่ำก็จะอ้วนง่าย ระบบการเผาส์พานไม่ค่อยดีเท่าไหร่ สำหรับคนที่เป็นโรคไทรอยด์ที่จริงแนะนำอยากให้กินเป็น Animal base มากกว่า เพราะว่าหนึ่งคุณต้องกินปริมาณที่ถึงต่อวัน อ่ะแน่นอนเรื่องของการออกกำลังกาย การที่กิน Animal base ด้วยความว่ามันมีกรดอามิโนที่เรียกว่าไทโรซีน กรดตัวนี้เป็นสารตั้งต้นของโฮโมนไทรอยด์ อ่า พอเรา พอคนไข้ตอนแรกอุ๊ย อยากลดน้ำหนัก ก็ไปกินเป็นแพนสเบส เพราะมันเป็นอะไรที่ฮิตฮอต แต่ระบบไทรอยด์เจ๊ง อ่า กะหรี่อ้วน ก็เลยอ้วน อ้วนกว่าเดิม หรือไม่ก็แบบพัง แล้วก็ระบบไทรอยด์ไม่ดีพัง แล้วคราวนี้ก็เลยเอ้ยลองปรับเปลี่ยนมาเป็น Animal แบบโปรตีนทั่วไปจากสัตว์ว์บ้างด้วยและกินถึงด้วยเป็น High Protein Diet พอปรับเปลี่ยนปุ๊บระบบการไทรอยด์ก็ทำงานปกติดีขึ้นซะด้วยซ้ำ อืออันนี้คือคนที่อยากน้ำหนักและถ้าคนที่อยากสร้างกล้ามเหนือ กล้ามเหนือก็ต้องกินเป็นแบบแอนิมัลแล้วเพราะแอนิมัลอ่ะให้จำนวนกล้ามโปรตีนที่เยอะกว่า ก็คือแปลว่าจริง จริงแล้วการเราเลือกกินโปรตีนอ่ะต้องเลือกก่อนว่าหนึ่ึ่งเราอยากกินเพราะอะไรใช่ป่ะ แล้วก็สองเราก็ต้องดูว่าร่างกายเราเป็นแบบไหนด้วย ถูกต้อง อือ อีกอันนึงที่อยากคุยเกี่ยวกับการลดน้ำหนักอ่ะ ก็คือเราพูดมาหลายแบบแล้วใช่ว่าเมื่อกี้พูดถึงเรื่องการกินโปรตีนเนาะ อันนี้ก็เลยอยากพูดเกี่ยวกับการงดอาหารหรือฟาสติ้งบ้าง ก็คือตั้งแต่เราเด็กอ่ะ
